```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
```

```python
data = pd.read_csv('iris.txt', sep='\t')
```

```python
cat = {'setosa' : 0, 'versicolor' : 1, 'virginica' : 2} 
data['Species'].replace(cat, inplace=True)
data.head()
```

```python
cmap = ListedColormap(['red', 'green', 'blue'])

fig, axs = plt.subplots(4, 4, figsize=(12, 12), layout='tight')

# Set x labels
axs[3, 0].set_xlabel('Sepal Length')
axs[3, 1].set_xlabel('Sepal Width')
axs[3, 2].set_xlabel('Petal Length')
axs[3, 3].set_xlabel('Petal Width')

# Set y labels
axs[0, 0].set_ylabel('Sepal Length')
axs[1, 0].set_ylabel('Sepal Width')
axs[2, 0].set_ylabel('Petal Length')
axs[3, 0].set_ylabel('Petal Width')

# Hide x and y labels subplots
for i in range(4):
    axs[0, i].set_xticks([])
    axs[1, i].set_xticks([])
    axs[2, i].set_xticks([])

    axs[i, 1].set_yticks([])
    axs[i, 2].set_yticks([])
    axs[i, 3].set_yticks([])

axs[0, 0].hist((data['Sepal.Length'].values))
axs[1, 0].scatter((data['Sepal.Length'].values), (data['Sepal.Width'].values), c=data['Species'].values, cmap=cmap)
axs[2, 0].scatter((data['Sepal.Length'].values), (data['Petal.Length'].values), c=data['Species'].values, cmap=cmap)
axs[3, 0].scatter((data['Sepal.Length'].values), (data['Petal.Width'].values), c=data['Species'].values, cmap=cmap)

axs[0, 1].scatter((data['Sepal.Width'].values), (data['Sepal.Length'].values), c=data['Species'].values, cmap=cmap)
axs[1, 1].hist((data['Sepal.Width'].values))
axs[2, 1].scatter((data['Sepal.Width'].values), (data['Petal.Length'].values), c=data['Species'].values, cmap=cmap)
axs[3, 1].scatter((data['Sepal.Width'].values), (data['Petal.Width'].values), c=data['Species'].values, cmap=cmap)

axs[0, 2].scatter((data['Petal.Length'].values), (data['Sepal.Length'].values), c=data['Species'].values, cmap=cmap)
axs[1, 2].scatter((data['Petal.Length'].values), (data['Sepal.Width'].values), c=data['Species'].values, cmap=cmap)
axs[2, 2].hist((data['Petal.Length'].values))
axs[3, 2].scatter((data['Petal.Length'].values), (data['Petal.Width'].values), c=data['Species'].values, cmap=cmap)

axs[0, 3].scatter((data['Petal.Width'].values), (data['Sepal.Length'].values), c=data['Species'].values, cmap=cmap)
axs[1, 3].scatter((data['Petal.Width'].values), (data['Sepal.Width'].values), c=data['Species'].values, cmap=cmap)
axs[2, 3].scatter((data['Petal.Width'].values), (data['Petal.Length'].values), c=data['Species'].values, cmap=cmap)
axs[3, 3].hist((data['Petal.Width'].values))

plt.show()
```
